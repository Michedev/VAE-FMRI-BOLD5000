```python
def check_all_data():
    roi = Path('data/ROIs')
    for cs in roi.dirs():
        print(cs.basename())
        if not (cs / 'mat').exists():
            print('  no mat')
            continue
        cs =  cs / 'mat'
        for mat in cs.files('*.mat'):
            print('  ', mat.basename())
            data = loadmat(mat)
            for k in data:
                if hasattr(data[k], 'shape'):
                    print('    ', k)
                    print('    ', data[k].shape)
```

```python
check_all_data()
```
